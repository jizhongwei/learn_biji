```python
import requests
from lxml import etree

url = "https://www.zdic.net/zd/zb/ty/bh/?bh=%s"
# req = requests.get(url)
# xp = etree.HTML(req.text)
# result = xp.xpath("//div[@class='bs_index3']/li/a/text()")
# print(dir(result))
# print(len(result))
# print(req.text)
# print(result)
def get_all(url):
    with open("all2.txt", 'w', encoding= 'utf-8') as fd:
        for i in range(5):
            url_to = url % str(i)
            req = requests.get(url_to)
            req.encoding = 'utf-8'
            xp = etree.HTML(req.text)
            result = result = xp.xpath("//div[@class='bs_index3']/li/a/text()")
            for r in result:
                fd.write(r + "\t")
    print("done!")

get_all(url)
```
